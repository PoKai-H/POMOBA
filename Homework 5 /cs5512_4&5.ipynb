```python


class SimpleMDP:
    def __init__(self, states, actions, transitions, terminal_states):
        self.states = states
        self._actions = actions
        self._transitions = transitions
        self.terminal_states = terminal_states

    def actions(self, s):
        return self._actions[s]

    def transitions(self, s, a):
        return self._transitions[(s, a)]

    def sample_transition(self, s, a):
        probs = self._transitions[(s, a)]
        r = random.random()
        cumulative = 0
        for s_next, p in probs:
            cumulative += p
            if r <= cumulative:
                return s_next
        return probs[-1][0]

    def is_terminal(self, s):
        return s in self.terminal_states
```

```python
def td_learning(mdp, episodes=5000, alpha=0.1, gamma=0.99, step_cost=-0.1):
    V = {s: 0 for s in mdp.states}

    for _ in range(episodes):
        s = random.choice(mdp.states)

        while not mdp.is_terminal(s):
            a = random.choice(mdp.actions(s))
            s_next = mdp.sample_transition(s, a)

            # rewards
            if s_next == s:
                r = -1  # wall penalty
            elif mdp.is_terminal(s_next):
                r = 1 if s_next == 'goal' else -1
            else:
                r = step_cost

            # TD update
            V[s] += alpha * (r + gamma * V[s_next] - V[s])

            s = s_next

    return V
```

```python
def get_reward(s, s_next, cost):
    if s == s_next: return -1.0 # Wall bump penalty
    if s_next == 'goal': return 1.0
    if s_next == 'bad': return -1.0
    return cost
```

```python
def extract_policy(V, states, actions, P, R, gamma):
    policy = {}
    for s in states:
        action_values = {}
        for a in actions:
            total = 0
            for s_next, prob in P[s][a].items():
                reward = R[s][a][s_next]
                total += prob * (reward + gamma * V[s_next])
            action_values[a] = total

        best_action = max(action_values, key=action_values.get)
        policy[s] = best_action

    return policy
```

```python
import numpy as np
import random

def generate_samples(num_samples=1000, P_ss=None, obs_matrix=None,
                      states=None, obs_list=None, start_state=0):

    samples = []

    s = start_state  # initial state

    for _ in range(num_samples):

        # ---- Sample next state from P(s' | s) ----
        probs = P_ss[:, s]

        # safety check (avoid invalid distributions)
        if np.sum(probs) == 0:
            s_next = start_state
        else:
            s_next = np.random.choice(states, p=probs)

        s = s_next

        # ---- Sample observation from P(y, a | s) ----
        obs_probs = obs_matrix[:, s]

        if np.sum(obs_probs) == 0:
            y_a = random.choice(obs_list)
        else:
            idx = np.random.choice(len(obs_probs), p=obs_probs)
            y_a = obs_list[idx]

        samples.append((s, y_a))  # store (state, observation-action pair)

    return samples
```

# Baum-Welch Algorithm (HMM Learning)  
Transitions: $P((s', e-1) | (s, e), a)$. If $e=0$, transition to terminal state.[cite: 1]
Rewards: Recharge states $s_{charge}$ provide a positive reward and reset $e$ to $C$.[cite: 1]
Input:
- Observations O = (o1, o2, ..., oT)
- Number of states N
- Number of observation symbols M

Output:
- Transition matrix A
- Emission matrix B
- Initial distribution π

Steps:

1. Initialize A, B, π randomly

2. Repeat until convergence:

   E-step:
   - Compute forward probabilities α_t(i)
   - Compute backward probabilities β_t(i)

   - Compute:
     γ_t(i) = P(state i at time t | O)
     ξ_t(i,j) = P(state i → j at time t | O)

   M-step:
   - Update initial:
     π_i = γ_1(i)

   - Update transitions:
     A_ij = Σ_t ξ_t(i,j) / Σ_t γ_t(i)

   - Update emissions:
     B_i(k) = Σ_t γ_t(i) where o_t=k / Σ_t γ_t(i)

In the next few problems we will consider the chicken domain from hw3, but with the following modifications:  Chickens can exchange crowns like currency and form teams.  Each chicken has a new component called a *sack*, which provides a mechanism for team building. Each chicken is initialized with an empty sack. Crowns earned in contests are automatically deposited in their sacks at the end of each tournament,  but they can also be transferred and ownership can be shared.  While sacks stay with a chicken, sack ownership can be extended to other chickens with the consequence of dividing the crowns amongst the owners at the end of the game. Ownership can only be extended by the original owner. With these modifications, we can have chickens specialize into observer and battle roles and observers can share victory through co-owned sacks. The goal is to be amongst the set of chickens that are *owners* of most crowns at the end of $T$ tournaments.  This can be a single chicken or a set of chickens.

Chickens have three new actions: *ShareObservations, TransferCrowns, ExtendSackOwnership* which can be used after all contests are fought and all observations delivered (e.g. at the end of the environment step).

* *ShareObservations(Action=IdOfChickenToShareWith)*: this action shares the chicken's observation history with another designated chicken.  

* *TransferCrowns(Action=(IdOfChickenToTransfer,NumberofCrowns)*:  Removes NumberofCrowns from chicken's own sack and puts them in the sack of the chicken with the id: *IdOfChickenToTransfer*.

* *ExtendSackOwnership(Action=(IdOfChickenToTransfer)* The action adds an owner to a chicken's sack (asymmetrically).  All crowns in that sack are divided among the owners at the end of the game to determine winners. For example if Chicken1 shares their sack of 120 crowns with Chicken2 and Chicken3, and Chicken2 shares their sack of 60 crowns with Chicken1 and Chicken3 shares their sack of 40 crowns with Chicken1, then Chicken1 get credited with 120/3+60/2+40/2=90, Chicken2 with 120/3+60/2=70, and Chicken3 with 120/3+40/2=60, so that Chicken1 is the sole winner at the end despite sharing.  If all three chickens had shared with each other, all would have won together.  The extendownership action is done without communication or guarantee or reciprocity.
